### Step 1: Cleaning recieved input - Cleaning

```python
def clean_text(text):
    for char in "!?.,;:'\"":
        text = text.replace(char, "")
    text = text.split()
    text = [word.lower() for word in text]
    return text

text= "Hello, World! This is a test. Are you ready?"
comments_clean = clean_text(text)
print(comments_clean)
```

### Step 2: Collect the unique set of words - Deduplicated and Flattened

```python
def build_vocab(comments_clean):
    vocab=[]
    #handling single comment only list by wrapping it which maintains the structure as lists of lists
    if isinstance(comments_clean[0],str):
        comments_clean=[comments_clean]
    for sentence in comments_clean:
        for word in sentence:
            if word not in vocab:
                vocab.append(word)
    return vocab

vocab = build_vocab(comments_clean)
print(vocab)
```

### Step 3: Index the words in vocab - Indexed

```python
def build_index(vocab):
    word_to_index={}
    count=0
    for word in vocab:
        if word not in word_to_index:
            word_to_index[word]=count
            count+=1
    return word_to_index

word_to_index = build_index(vocab)
print(word_to_index)
```

### Step 4: Run end to end - Tokenizer Orchestrator

```python
def tokenizer(comments):
    comments_clean=[]
    for comment in comments:
        comments_clean.append(clean_text(comment))
    print(comments_clean)
    comments_vocab = build_vocab(comments_clean)
    print(comments_vocab)
    vocab_indexed = build_index(comments_vocab)
    print(vocab_indexed)

comments = [
    "You are so STUPID!!",
    "This is a wonderful day, isn't it?",
    "I hate you so much!!!",
    "Wonderful people make the world better.",
    "Stupid, hate-filled comment right here?"
]

tokenizer(comments)
```
