```python
def clean_text(comments):
    if isinstance is (comments[0],str):
        comments = [comments] #handles single comment edge case
    cleaned_comments= []
    for comment in comments:
        for char in "?!',;\"":
            comment=comment.replace(char,"")
        comment = comment.lower()
        comment = comment.split()
        cleaned_comments.append(comment)
    return cleaned_comments
        

comments=["I really like this song!", "Is it even worth my time?", "This is amazing stuff!!"]
cleaned_comments= clean_text(comments)
print(cleaned_comments)
```

```python
def build_vocab(cleaned_comments):
    vocab=[]
    for sentence in cleaned_comments:
        for word in sentence:
            if word not in vocab:
                vocab.append(word)
    return vocab

vocab = build_vocab(cleaned_comments)
print(cleaned_comments)
```
